### Sklearn reference

http://scikit-learn.org/stable/modules/tree.html

```python
from sklearn import tree
```

```python
X = [[0, 0], [1, 1]]
```

```python
X
```

```python
Y = [0, 1]
```

```python
clf = tree.DecisionTreeClassifier()
```

```python
clf = clf.fit(X, Y)
```

```python
pred = clf.predict([[2., 2.]])
pred
```
